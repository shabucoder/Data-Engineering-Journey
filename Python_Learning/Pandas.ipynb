GROUPBY - The Foundation


**What is GroupBy?**
- GroupBy is like organizing data into groups and then performing calculations on each group.

_Real-life analogy:_

- Imagine you have sales data from multiple stores
- You want to find total sales per store
- GroupBy groups all transactions by store, then sums them up

```python
#### Grpupby in pandas dataFrame 

import pandas as pd 
import numpy as np 

data = {
    'Stores':['storeA', "storeB", 'storeA','storeC', 'storeB', 'storeA', 'storeC',],
    'Product':['Laptop','Phone','Tablet', 'Laptop', 'Phone','Tablet','Tablet'],
    'Sales': [1200, 800, 600,1500,900,700,550],
    'Quantity':[2,4,3,3,5,2,4]

}

df = pd.DataFrame(data)
df.head(7)



```

```python
### To find total sales per store

total_sales_store = df.groupby('Stores')['Sales'].sum()
print(total_sales_store)
```

```python
### B) Multiple Aggregations

store_status = df.groupby('Stores')['Sales'].agg(['sum', 'count', 'mean'])
print(store_status)
```

```python
### Group by Multiple Columns

store_sales_per_product = df.groupby(['Stores','Product'])['Sales'].sum()
print(store_sales_per_product)

store_sales_per_product_count = df.groupby(['Stores','Product'])['Sales'].agg(['sum','count'])
print(store_sales_per_product_count)
```

**Different Aggregation Functions**

```python
## Multiple columns, multiple aggregations

result = df.groupby('Stores').agg({'Sales':['sum','mean',"max", "min"], 'Quantity':['sum', 'mean']})
print(result)
```

**PIVOT TABLES - Advanced Summarization**

_What is a Pivot Table?_
- A Pivot Table is like a smart cross-tabulation that reshapes data and summarizes it.
Real-life analogy:

- You have sales data with Store, Product, and Month
- You want a table showing Sales per Product (rows) for each Month (columns)
- Pivot table creates this summary automatically!

```python
##Sample Data
sales_data = pd.DataFrame({
    'Date': ['2024-01', '2024-01', '2024-02', '2024-02', '2024-01', '2024-02'],
    'Product': ['Laptop', 'Laptop', 'Laptop', 'Phone', 'Tablet', 'Tablet'],
    'Store': ['Store A', 'Store A', 'Store A', 'Store A', 'Store B', 'Store B'],
    'Sales': [1200, 800, 1500, 900, 600, 700]
})

print(sales_data)
print("\n")
```

```python
## You want a table showing Sales per Product (rows) for each Month (columns)

pivot = sales_data.pivot_table(
    values='Sales',
    index= 'Product',
    columns='Date',
    aggfunc='sum'
)

print(pivot)
```

```python
pivotb = sales_data.pivot_table(
    values= 'Sales',
    index = ['Store', 'Product'],
    columns= 'Date',
    aggfunc = 'sum',
    fill_value= 0
)

print(pivotb)
```

```python
### Real Data Engineering Example - Complete Analysis

import numpy as np

transactions = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=20,freq='D').tolist()*2,
    'Category':['Electronics', 'Clothing']*20,
    'Product': ['Laptop', "shirt", 'Phone', 'Pants', 'Tablets', 'Jackets']*6 +['Laptop', 'Shirt']*2,
    'Store':['StoreA','StoreB']*20,
    'Sales':np.random.randint(100,1000,40),
    'Quantity':np.random.randint(1,10,40)
})
```

```python
print(transactions.head(14))
print("\n")

print('='*50)
print('Groupby Analysis')
print('='*50)
print('\n')
```

```python
print('='*50)
print('1. Total Sales per category')
print('='*50)

salesPercategory = transactions.groupby('Category')['Sales'].sum()
print(salesPercategory)

print("\n2. Average Sales per Store:")
print('='*50)

AvgSalesPerStore = transactions.groupby('Store')['Sales'].mean()
print(AvgSalesPerStore)

print("\n3. Detailed Stats per Category:")
print('='*50)

stats_category = transactions.groupby('Category').agg({
    'Sales':['sum', 'mean', 'count'],
    'Quantity':['sum']})

print(stats_category)

```

```python
## PIVOT TABLE Analysis
print("\n" + "=" * 50)
print("PIVOT TABLE ANALYSIS")
print("=" * 50)

## 1. Sales by Category and Store
print("\n1. Sales by Category (rows) and Store (columns):")

pivotA = transactions.pivot_table(
    values = 'Sales',
    index = 'Category',
    columns = 'Store',
    aggfunc = 'sum'
)

print(pivotA)

## 2. Average sales by Product and Store
print("\n2. Average Sales by Product (rows) and Store (columns):")
pivot2 = transactions.pivot_table(
    values='Sales',
    index='Product',
    columns='Store',
    aggfunc='mean'
)
print(pivot2.round(2))
```

**Common GroupBy & Pivot Patterns in Data Engineering**

```python
transactions['Date'] = pd.to_datetime(transactions['Date'])
transactions.info()

df= transactions.groupby(transactions['Date'].dt.date)['Sales'].sum()
print(df)
```

```python
transactions['mod_Date'] = transactions['Date'].dt.date
print(transactions)
```

```python
### Top N per group 

top_products = transactions.groupby('Category').apply(
    lambda x:x.nlargest(3,'Sales')
)

print(top_products)
```

```python
## Cumulative sum per group 

transactions['cumulative_Sales'] = transactions.groupby('Store')['Sales'].cumsum()
print(transactions)
```
