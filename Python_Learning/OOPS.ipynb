`What is a Class?`
- A class is a **blueprint or template** for creating objects. Think of it like:

`Examples:`
- A cookie cutter (class) that makes cookies (objects)
- A car blueprint (class) that creates actual cars (objects)
- A house plan (class) that builds actual houses (objects)

`What is an Object?`
- An object is an instance (actual copy) created from a class.

__If Class = Recipe, then Object = Actual cake you baked

`What are Instance Variables?`

- Variables that **belong to a specific object (instance)**. Each object has its own copy.

***2. Inheritance (Child classes inherit from Parent)***
- Definition: A new class can inherit properties and methods from an existing class

```python
## Single Inheritence
## Creating the Parent Class(Base Class)
class Vehicle():
    def __init__(self,name,price):
        self.name = name
        self.price = price

    def start(self):
        print(f"{self.name} is starting")
    def stop(self):
        print(f"{self.name} has stopped")

## Creating child class and inheriting the base class inside the child class 
class Bike(Vehicle):
    def __init__(self, name, price, gastype):
        super().__init__(name, price)        ##Ingeriting the parent class with constructor inside 
        self.gastype = gastype 

    def typeofGas(self):
        print(f"{self.name} is {self.gastype} vehicle")
```

```python
bike24 = Bike('Honda', '24K', 'Petrol')

bike24.start()
bike24.typeofGas()
```

```python
### Multiple Inheritance 
### When class is inherits from  more than one class
## creating the Animal Class 

## First base class
class Animal():
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("Subclass must implement this method")

    def legs(self):
        print(f"{self.name} has four legs")


class Pet():
    def __init__(self, Owner):
        self.Owner = Owner


class Dog(Animal, Pet):
    def __init__(self,name,Owner, age):
        Animal.__init__(self,name)
        Pet.__init__(self, Owner)
        self.age = age

    def speak(self):
        print(f"{self.name} says whoof whoof!")
        print(f'{self.name} age is {self.age}')




dog1 = Dog('Buddy', 'Krish', 24)
dog1.speak()
print(f"Owner:{dog1.Owner}")












```

```python
#### Polymorphism 

class Dog():
    def speak(self):
        return "whoof whoof!"
    
class cat:
    def speak(self):
        return "Meow Meow!"
    
class Cow:
    def speak(self):
        return "Moo!"
    
animals = [Dog(), cat(), Cow()]

for animal in animals:
    print(animal.speak())

    
```

```python
### Method overridding


class BankAccount():
    def __init__(self, name, balance=0):
        self.name = name 
        self.balance = balance 

    def withdraw(self, amount):
        if self.balance >=amount:
            self.balance -=amount
            print(f"Withdrew:{amount}")
        else:
            print("Insufficient funds")

class SavingsAccount(BankAccount):
    def __init__(self, name, balance=0, interest_rate = 0.02):
        super().__init__(name, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest added: {self.balance}")

class CurrentAccount(BankAccount):
    def __init__(self, name , balance, overdraft_limit = 500):
        super().__init__(name, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):   ### Method overridding from parent class
        if self.balance + self.overdraft_limit >= amount:
            self.balance -=amount
            print(f"withdrew:${amount}")
        else:
            print("Exceeds overdraft limit")


##Usage 

savings = SavingsAccount("Alice", 1000)
savings.add_interest()

current = CurrentAccount("Bob", 100)
current.withdraw(500)

Account = BankAccount("bob", 5000)
Account.withdraw(500)





        
```

```python
class Employee():   ### Parent Class
    def __init__(self, name, employee_id, base_salary=0):
        self.name = name
        self.employee_id = employee_id
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary

    def display_info(self):
        print(f"employee Name: {self.name}")
        print(f"Employee ID: {self.employee_id}")
        print(f"Base Salary: {self.base_salary}")

class Manager(Employee):
    def __init__(self, name, employee_id, base_salary=0, team_size=0):
        super().__init__(name, employee_id, base_salary)
        self.team_size = team_size

    def calculate_salary(self):
        return self.base_salary + (self.team_size * 500)
    
class Developer(Employee):
    def __init__(self, name, employee_id, base_salary=0, ProgrammingLanguage = None, Projects_completed=None):
        super().__init__(name, employee_id, base_salary)
        self.programmingLanguage = ProgrammingLanguage
        self.Projects_completed = Projects_completed

    def calculate_salary(self):
        return self.base_salary +(self.Projects_completed * 200)
    

class Intern(Employee):
    def __init__(self, name, employee_id, base_salary=0, university=None):
        super().__init__(name, employee_id, base_salary)
        self.university = university

    def calculate_salary(self):
        intern_salary = self.base_salary*0.5
        return intern_salary



employees = [Manager('Sakshith', 1, 40000, 4), Developer('Chai', 2, 35000, 'Python', 4), Intern("Rohith", 3, 35000, 'TXST')]


for employee in employees:
    Final_salary= employee.calculate_salary()
    print(f"{employee.name}:{Final_salary}")


```
